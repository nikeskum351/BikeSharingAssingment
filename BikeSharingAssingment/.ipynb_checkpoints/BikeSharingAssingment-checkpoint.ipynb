## Data Cleaning

```python
# droping instant as its just index
# droping dteday column as it is have yr, mnth and weekday columns
# droping casual and registered as cnt is sum of (casual + registered)
day_data_frame = day_data_frame.drop(['instant', 'dteday','casual','registered'], axis=1)
day_data_frame.head()
```

```python
# renaming few columns for better readibility

day_data_frame.rename(columns={'yr':'year','mnth':'month','hum':'humidity'}, inplace=True)
day_data_frame.head()
```

```python
#checking for null or missing values
day_data_frame.isnull().sum()
```

```python
#converting seasons to categorical variable
day_data_frame['season'] = day_data_frame['season'].map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
day_data_frame.head()
```

```python
#converting weathersit to categorical variable
day_data_frame['weathersit'] = day_data_frame['weathersit'].map({1:'Clear', 2:'Mist_Cloudy', 3:'Light_Rain_Snow', 4:'Heavy_Rain_Snow'})
day_data_frame.head()
```

```python
#converting month to categorical variable
day_data_frame['month'] = day_data_frame['month'].map({1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'june',7:'july',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'})
day_data_frame.head()
```

```python
#converting weekday to categorical variable
day_data_frame.weekday = day_data_frame.weekday.map({0:'sun',1:'mon',2:'tue',3:'wed',4:'thu',5:'fri',6:'sat'})
day_data_frame.head()
```

```python
day_data_frame.info()
```

## Data Preparation

```python
# Get the dummy variables for the feature 'season' and store it in a new variable - 'season_new'
season_new = pd.get_dummies(day_data_frame['season'], dtype=int, drop_first=True)
```

```python
season_new.head()
```

```python
# Get the dummy variables for the feature 'weathersit' and store it in a new variable - 'weathersit_new'
weathersit_new = pd.get_dummies(day_data_frame['weathersit'], dtype=int)
weathersit_new.head()
```

```python
# Get the dummy variables for the feature 'weekday' and store it in a new variable - 'weekday_new'
weekday_new = pd.get_dummies(day_data_frame['weekday'], dtype=int, drop_first=True)
weekday_new.head()
```

```python
# Get the dummy variables for the feature 'month' and store it in a new variable - 'month_new'
month_new = pd.get_dummies(day_data_frame['month'], dtype=int, drop_first=True)
month_new.head()
```

```python
# Add the dummy columns to day_data_frame
day_data_frame = pd.concat([day_data_frame, month_new, weekday_new, weathersit_new, season_new], axis=1)
day_data_frame.head()
```

```python
# Drop 'season', 'month' and 'weathersit' as we have created the dummies for it
day_data_frame.drop(['season', 'weathersit', 'weekday','month'], axis=1, inplace= True)
day_data_frame.head()
```

```python
day_data_frame.info()
```

## Splitting the Data into Training and Testing Sets

```python
#Splitting the Data into Training and Testing Sets
df_train, df_test = train_test_split(day_data_frame, train_size=0.7, test_size=0.3, random_state=100)
```

## Feature Scaling continuous variables

```python
# Resacaling the features
scaler = MinMaxScaler()
num_vars = ['temp', 'atemp', 'humidity', 'windspeed', 'cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
#Dividing into X and Y sets for the model building
y_train = df_train.pop('cnt')
X_train = df_train
y_train.head()
```

```python
X_train.head()
```

```python
# Running RFE with the output number of the variable equal to 17
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(estimator=lm, n_features_to_select=16)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

# Model1 using statsmodel, for the detailed statistics

```python
#Building model using statsmodel, for the detailed statistics
X_train_rfe_sm = sm.add_constant(X_train_rfe)
```

```python
 # Running the linear model
lm = sm.OLS(y_train,X_train_rfe_sm).fit()  
```

```python
print(lm.summary())
```

```python
# Calculate the VIFs for the  model1
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Column Manual Elimination based on p value and VIF
   - High p-vale, high VIF
   - High-low:
        - High p, low VIF : remove these first
        - Low p, high VIF : remove these after the ones above
   - Low p, low VIF

```python
# Dropping highly correlated variables and insignificant variables
#remove first High p, low VIF --- so dropping the dec column
X_train_rfe = X_train_rfe.drop('dec', axis=1)
X_train_rfe.head()
```

## Making Predictions

```python
# Applying the scaling on the test sets
# Resacaling the features
num_vars = ['temp', 'atemp', 'humidity', 'windspeed', 'cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.head()
```

#### Dividing into X_test and y_test

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
y_test.head()
```

```python
X_test.head()
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_rfe.columns]

# Adding a constant variable 
X_test_new_sm = sm.add_constant(X_test_new)
X_test_new_sm.head()
```

```python
# Making predictions
y_pred = lm.predict(X_test_new_sm)
```

 ## Model Evaluation

```python
# Plotting y_test and y_pred to understand the spread.

fig = plt.figure()
plt.scatter(y_test, y_pred)
plt.suptitle("y_test vs y_pred", fontsize=20)
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
plt.show()
```

#### Checking the R-squared on the test set and comparing with train set

```python
# r2 score of train set
r2_train = r2_score(y_train, y_train_cnt)
r2_train
```

```python
# r2 score of test set
r2_test = r2_score(y_test, y_pred)
r2_test
```

```python
# Calculating Adjusted-R^2 value for the test dataset

adjusted_r2_test = round(1-(1-r2_test)*(X_test_new.shape[0]-1)/(X_test_new.shape[0]-X_test_new.shape[1]-1),4)
print(adjusted_r2_test)
```

- Train dataset R^2          : 0.835
- Test dataset R^2           : 0.795
- Train dataset Adjusted R^2 : 0.832    
- Test dataset Adjusted R^2  : 0.785

```python
# Visualizing the fit on the test data
# plotting a Regression plot

plt.figure()
sns.regplot(x=y_test, y=y_pred, ci=68, fit_reg=True,scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)
plt.show()
```
